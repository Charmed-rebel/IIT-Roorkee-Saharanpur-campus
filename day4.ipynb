```python
grades = [tuple(),('roll no.1',78),('roll no. 2',90),('roll no. 3',64),('roll no. 4',34),('roll no. 5',70)]
grades.append(('roll no. 6', 88))
def marks(n): 
    return grades[n][1]
n = int(input('enter the roll no. of student: '))
print(f"Marks of student with roll no. {n}: {marks(n)}")
```

```python
names = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']
sorted_names = sorted(names)                               # alphabetically 
print("Sorted names:", sorted_names)
```

```python
names = ['Alice', 'Bob', 'Charlie', 'David', 'Eve']
sorted_names = sorted(names, reverse=True)                  # reverse alphabetically
print("Sorted names:", sorted_names)
```

```python
A = [[12, 92, 83], [14, 95, 76], [67, 22, 13]]
B = [[97, 28, 67], [69, 55, 34], [33, 12, 19]]
if len(A) != len(B) or len(A[0]) != len(B[0]):
    print("Error: Matrices must have the same dimensions.") 
else:
    result = []
    for i in range(len(A)):                       
        row = []
        for j in range(len(A[0])):
            row.append(A[i][j] + B[i][j])
        result.append(row)
print(result)   
```
