```python
class BankAccount:
    def __init__(self, owner_name, starting_balance=0.0):
        self.owner = owner_name
        self.balance = starting_balance
        print(f"Account created for {self.owner} with starting balance: ${self.balance}")

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount}. Current balance: ${self.balance}")
        else:
            print("Failed: Deposit amount must be greater than zero.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ${amount}. Current balance: ${self.balance}")
        else:
            print("Failed: Insufficient funds or invalid amount.")

    def display_balance(self):
        print(f"Account Balance for {self.owner}: ${self.balance}")
        return self.balance
my_account = BankAccount("Alice", 100.0)

my_account.deposit(50.0)
my_account.withdraw(20.0)
my_account.withdraw(500.0)
my_account.display_balance()
```

```python
import pandas as pd

dummy_data = """Name,Age,Score
Alice,24,88.5
Bob,,92.0
Charlie,30,
David,22,79.0"""
with open("student_data.csv", "w") as file:
    file.write(dummy_data)
df = pd.read_csv("student_data.csv")
print("Original Data")
print(df)
print("\ndf.info(): Technical details about the columns")
df.info()
print("\n df.describe(): Instant Math/Stats on numerical columns")
print(df.describe())
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Score'] = df['Score'].fillna(0)

print("\n--- Cleaned Data (No more missing values!) ---")
print(df)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank'],
    'Major': ['Math', 'Physics', 'Math', 'Physics', 'Math', 'Biology'],
    'Score': [85, 92, 78, 88, 95, 80]
}
df = pd.DataFrame(data)
print("--- Full Dataset ---")
print(df)
high_scorers = df[df['Score'] > 80]

print("\n--- High Scorers (Score > 80) ---")
print(high_scorers)
average_by_major = df.groupby('Major')['Score'].mean()

print("\n--- Average Score per Major ---")
print(average_by_major)
average_by_major.plot(kind='bar', color='skyblue', edgecolor='black')

plt.title('Average Score by Academic Major')
plt.xlabel('Major')
plt.ylabel('Average Score')
plt.show()
```

```python
class Book:
    def __init__(self, title):
        self.title = title
        self.is_borrowed = False 
class Member:
    def __init__(self, name):
        self.name = name
        self.borrowed_books = []

    def borrow_book(self, book):
        if book.is_borrowed == False:
            book.is_borrowed = True
            self.borrowed_books.append(book.title)
            print(f"Success! {self.name} borrowed '{book.title}'.")
        else:
            print(f"Sorry, '{book.title}' is already checked out.")
book1 = Book("Field Theory and Roots")
book2 = Book("Learning Braille Step-by-Step")
member1 = Member("Alice")
member1.borrow_book(book1)
member1.borrow_book(book1)
```

```python
import numpy as np
matrix_A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])
matrix_B = np.array([
    [7, 8],
    [9, 10],
    [11, 12]
])

print(" Matrix A (2x3)")
print(matrix_A)

print("\n Matrix B (3x2)")
print(matrix_B)
result_matrix = matrix_A @ matrix_B
print("\nResulting Matrix (2x2)")
print(result_matrix)
```

```python
import numpy as np
matrix_A = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

matrix_B = np.array([
    [7, 8],
    [9, 10],
    [11, 12]
])
result_1 = np.dot(matrix_A, matrix_B)
result_2 = matrix_A.dot(matrix_B)

print("--- Result using np.dot(A, B) ---")
print(result_1)

print("\n--- Result using A.dot(B) ---")
print(result_2)
```

```python
import pandas as pd
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
titanic_df = pd.read_csv(url)

print(" Successfully loaded the Titanic Dataset!")
print(f"Total Passengers Analyzed: {len(titanic_df)}\n")
print("First 5 Rows ")
print(titanic_df.head())
median_age = titanic_df['Age'].median()
titanic_df['Age'] = titanic_df['Age'].fillna(median_age)
survival_rate = titanic_df['Survived'].mean() * 100
print(f"\nOverall Survival Rate: {survival_rate:.2f}%")
gender_survival = titanic_df.groupby('Sex')['Survived'].mean() * 100
print("\n--- Survival Rate by Gender ---")
print(gender_survival)
class_survival = titanic_df.groupby('Pclass')['Survived'].mean() * 100
print("\n--- Survival Rate by Passenger Class ---")
print(class_survival)
```
